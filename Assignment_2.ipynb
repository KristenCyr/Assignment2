#### Distribution Expansion of the Blacklegged Tick (*Ixodes scapularis*) Causing an Increase of Lyme Disease Cases in Nova Scotia, Canada.

#### *Introduction*

   The black legged tick (*Ixodes pacificus*) is a hard-bodied arthropod (Gabriele-Rivet et al. 2015). Its morphology is like many species of ticks making it difficult to differentiate. They are most commonly identified microscopically using morphological characteristics (Governement du Quebec 2019).    
   
   The species *I. pacificus* are found primarily on the West coast (Government of Canada 2017). In recent years, *I. pacificus* have been extending their range across Canada into Eastern regions (Government of Canada 2017). Previous studies have found geographic distribution patterns of these ticks are directly linked to temperature patterns (Levy et al. 2017; Hatchette et al. 2013; Gabrielle; Burtis et al. 2016). These patterns include higher activity during warmer summer months between June-August, a preference for temperatures over 7<sup>0</sup>C and humidity over 80% (Macdonald et al. 2016). In addition to temperature *I. pacificus* require forested habitats and an assortment of vertebrate hosts (Levy et al. 2017; Government of Canada 2017).
   
   Like other ticks, *I. pacificus* have three life stages including the  larvae, nymph and adult (Levy 2017). A hemoglobin diet is essential for the development through these stages (Levy et al 2017). The primary hosts during larvae and nymph stages are white-footed mice, and other small rodents (Levy et al. 2017). White-tailed deer are the most common hosts for adult ticks (Levy et al. 2019). However, *I. pacificus* also attach to humans, dogs, cats and other larger vertebrates (Levy et al. 2019).
   
Unlike many species of ticks, *I. pacificus* carries a bacterium called *Borrelia burgdorferi* (Gabriele-Rivet et al. 2015; Kulkarni et al. 2019; Macdonald et al. 2016). The bacterium has been proven to cause a vector borne disease known as Lyme disease (LD) (Kulkarni 2019). This disease is primarily transmitted through a single bite of *I. Pacificus* and is the highest vector borne disease in North America (Kulkari et al. 2019).
    
When a tick attaches to a human the symptoms of LD occur 3-30 days after transmission (Government of Canada 2017). The symptoms can differ from person to person making it difficult to diagnose as LD (Government of Canada 2017). Additionally, mild symptoms of LD are common across other diagnoses, including fatigue, rash, or flu, and contributes to the difficulty of a diagnosis (Government of Canada 2017). These symptoms can be treated with antibiotics if detected early enough. However, if left untreated LD can cause permanent neurological and heart disorders, and facial paralysis (Government of Canada 2017). 
    
The increase of temperatures from climate change in the last 15 years have caused higher abundances of *I. pacificus* on the East coast of Canada (Levy et al 2017; Hatchette et al. 2015; Gabriele-Rivet et al. 2015; Macdonald et al. 2016). This suggests there is a higher risk of LD in Nova Scotia than previous years and further suggests the risk will continue rising (Gabriele-Rivet et al. 2015). Therefore, the purpose of this study is to identify the number of LD cases from 2005-2017 across Nova Scotia. Determining the number of LD cases over a 12 year period will allow us to identify the magnitude of risk to public and animal health. Additionally, including information on LD cases by region will help identify areas of high prevalence. Analyzing Nova Scotia regions of LD risk will also provide a baseline for predicting future effects of climate change on *I. pacificus* distribution. This will help build awareness of current and future risks to public health and could help inform management on LD prevention plans.

#### *Methods & Results*

Data in this study were exported from data Nova Scotia, https://data.novascotia.ca/stories/s/ijy9-q5bk. They included number of LD cases, both confirmed and probable, from 2005-2017 in the entire province of Nova Scotia and per region of Nova Scotia, distinguished as Central, Western, Eastern, and Northern. These zones were not clearly defined in the exported data set. The first data set was used to identify the total number of LD cases in Nova Scotia over 12 years and the second data set was used to show evidence of distributional patterns. Cases of LD were grouped into confirmed and probable to ensure this study was not over reporting actual cases. They were exported directly from the website as a csv file. These two data sets were imported into R and analyzed using additional libraries, dplyr, stringr, and ggplot2.

```r
#Import data

tick_data = read.csv("Lyme_Disease_Counts_and_Rates__Displays_Confirmed__Probable_.csv")
tickregion_data = read.csv("Lyme_Disease_Counts_and_Rates_by_Zone__Displays_Confirmed__Probable_.csv")
head(tick_data)
```

```r
#Packages used
library(dplyr)
library(stringr)
library(ggplot2)
```

The first data set was grouped together by year and disease. The grouping was used with the summary function to calculate a total count of probable and confirmed cases per year.  

```r
#Lyme Disease Cases Per Year

tick_data = tick_data %>% group_by(Year, Disease) %>%
  summarise(Total_Cases = sum(Number.of.Cases))
head(tick_data)
```

A new data frame was created to plot the total number of LD cases per year generated from the previous "tick_data" data frame. This data frame was formatted into a scatterplot with scientific writing standards using ggplot. Gridlines, and borders were removed, axis labels were properly named, and a caption and title were included and adjusted to fit the text in the image box. A linear regression model was used to create a line of best fit for both confirmed and probable cases. 


```r
#Scatterplot of tick_data
tick_plot = ggplot(tick_data, aes(fill = Disease, x = Year, y = tick_data$Total_Cases)) + 
  geom_point(size=2, shape=23) + 
  geom_smooth(method = "lm", se=FALSE, color="black", formula = y ~ splines::bs(x,3)) +
  ggtitle("Total Reported Lyme Disease Cases per Year in Nova Scotia") +
  scale_fill_manual("legend", values = c("Lyme Disease - Confirmed" = "red", "Lyme Disease - Probable" = "grey")) +
  scale_x_continuous(breaks = c(2005:2017)) + 
  theme_bw() + 
  theme(panel.border = element_blank(), panel.grid.major = element_blank(), panel.grid.minor = element_blank(), axis.line = element_line(colour = "black"), plot.caption = element_text(hjust = 0))

tick_plot + 
  labs(y = "Total Cases",
      caption = "Fig. 1  Total number of confirmed and probable cases of Lyme disease 
per year from 2005-2017. Linear model function was used to generate line of best fit. 
Data was exported from data Nova Scotia 2019 and graph was generated using ggplot.") 
```

There was an exponential increase of LD cases from 2005-2017 (Fig. 1). Confirmed cases were higher than probable cases but both showed a similar pattern (Fig. 1). Reported cases were the lowest from 2005-2010 (Fig.1) There was a steep increase in cases after 2010 (Figure 1). The highest number of total confirmed cases was in 2017, with a value of approximately 300 confirmed cases (Fig. 1). 

The dataframe, region_data, was grouped by zone, into probable and confirmed cases. The zones included Western, Central, Northern and Eastern. The grouped data was summarized to calculate the average number of cases by zone over the 12 years. The rows specifying the averages for all of Nova Scotia were redundant and therefore were removed using indexing. 

```r
#Calculating average Cases per Region
region_data = tickregion_data %>% group_by(Zone, Disease) %>%
  summarise(Average_Cases = mean(Number.of.Cases))

region_data = region_data[-c(7:8), ]

head(region_data)
```

A new data frame was created using "region_data". The data was plotted as a stacked barplot showing the average probable and confirmed cases from 2005-2017 across the 4 zones. The dataframe was formatted into scientific writing standards using ggplot. Gridlines, and borders were removed, axis labels were properly named, and a caption and title were included and adjusted to fit the text in the image box. 

```r
#Stacked barplot of Region_data
region_plot = ggplot(region_data, aes(fill = Disease, x = Zone, y = region_data$Average_Cases)) + 
  geom_bar(position="stack", stat="identity") + 
  ggtitle("Average Reported Lyme Disease Cases by Zone") +
  scale_fill_manual("legend", values = c("Lyme Disease - Confirmed" = "red", "Lyme Disease - Probable" = "grey")) +
  theme_bw() + 
  theme(panel.border = element_blank(), panel.grid.major = element_blank(), panel.grid.minor = element_blank(), axis.line = element_line(colour = "black"), plot.caption = element_text(hjust = 0))

region_plot + 
  labs(y = "Average Cases", 
       caption = "Fig. 2  Average number of confirmed and probable cases of Lyme disease 
in Central, Western, Eastern, and Northern Nova Scotia over 12 years. 
Data was from 2005-2017. Data was exported from data Nova Scotia 2019.") 
```

There was a pattern of decline in LD cases moving from Western to Eastern Nova Scotia (Fig. 2). Western Nova Scotia had the highest prevelance of LD (Fig. 2). The zone with the second highest average of cases was central Nova Scotia (Fig. 2). Northern and Eastern zones had the lowest amount of cases, however Eastern had the least prevalence (Fig.2). There were both probable and confirmed cases in all zones except Eastern Nova Scotia where all cases were confirmed (Fig. 2). 

A new data frame was created from using the dataframe "tick_data". The new data frame used the linear model function to calculate if there was a significant difference in the total number of cases comparing across the 12 years. The summary function on the new data frame, "linearmod", was used to generate the R<sup>2</sup> and p-value. 

```r
#Linear Model of Total Cases 
linearmod = lm(Total_Cases ~ Year, data = tick_data)
summary(linearmod)
```

 Total number of LD cases in Nova Scotia from 2005-2017 had a significant difference (F<sub>(1,24)</sub> = 48.65, p-value = 3.266e-07). The total number of cases increased with year (R<sup>2</sup> = 0.6559). Therefore, LD cases have significantly between 2005-2017.

#### *Discussion*

Results showed a significant exponential increase of LD cases from 2005-2017 with a higher prevelance in Western Nova Scotia (Fig.1; Fig. 2). This suggests as warming temperatures create more suitable habitats for *I. pacificus*, total LD cases will continue to rise. Additionally, *I. pacificus* will likely continue increasing their distribution across Nova Scotia causing an increase of LD cases in the Eastern zone. This raises critical conerns for the future of public and animal health in Nova Scotia.

Until 1990 there were no species of *I. pacificus* found in Canada (Leo et al. 2017). The species are primarily constrained by habitat (Sonenshine D. 2018; Leo et al. 2014). However, recent studies have provided evidence of significant changes to their habitat range as a result of climate change (Sonenshine D. 2018; Metcalfe Z. 2017; Burtis et al. 2016). Their expansion has increased 45km/year in Ontario and nearly four-fold across a six-year study in Quebec (Sonenshine D. 2018). Therefore, it is likely the distribution of *I. pacificus*  is undergoing a similar increase across Nova Scotia. Additionally, *I. pacificus* accounts for nearly 95% of LD cases (Kulkari et al. 2019). This supports results found for the exponential increase in cases across Nova Scotia from 2005-2017 (Fig. 1). Climate change will continue to increase the ditribution of *I. pacificus* and by proxy LD cases. Therefore, it is imperative better management practises are put in place to prevent exposure such as signage at the beginning of trails to indicate areas with higher tick abundances and how to prevent exposure (Metcalfe Z. 2017). Future studies should consider animal cases of LD to better understand the effects of LD on the ecosystem.

The geographic range expansion of *I. pacificus* is primarily via migratory birds and white-tailed deer (Sonenshine D. 2018; Leo et al. 2017; Levy S. 2017). When the vector of expansion is migratory birds there is a patchiness of abundance throughout the region (Leo et al. 2017). In contrast, patterns of total cases will show a gradual decrease  representative of the distribution range of white-tailed deer when the deer are used as the vector (Levy S. 2017). This pattern of decrease supports results found for LD cases in Western, central, and Eastern Nova Scotia. White-tailed deer cross through the Western Zone (Paterson et al. 2002). As resources become limited they travel across the central zone, and some continue to the East (Paterson et al. 2002). A similar trend of decrease in LD cases was seen from West to East.  However, this could also be attributed to the temperature and humidity requirements of the species. Eastern Nova Scotia is a suboptimal environment for *I. pacificus* in comparison to the Western region (Hatchette et al.2015d therefore LD cases are lower because populations can’t be sustained after distribution via migratory birds (Hatchette et al; Government of Canada 2017). Future studies should consider comparing cases over more specific regions that are clearly defined. This will help provide an understanding of the mechanism for range expansion of *I. pacificus* in Nova Scotia. A better understanding of this mechanism will help monitor and predict the spread of LD in the future as climate change creates more optimal habitats for *I. pacificus*.
